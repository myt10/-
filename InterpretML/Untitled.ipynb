```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open("result.txt") as f:
    a = f.readlines()
count = 0
for i in range(len(a)-1):
    count += int(a[i][0])
(count+1)/(len(a)-1)
```

```python
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
name_list = ['Linear Regression','Decision Tree','Decision Rule','LIME','SHAP(取1/4训练集)','EBMBoosting']
num_list = [7,4,8,139,661,51]
fig, ax = plt.subplots()

for a, b in zip(name_list, num_list):
    ax.text(a, b+1, b, ha='center', va='bottom',size = 15)
b = ax.bar(name_list, num_list,width = 0.35,color = ['b','deeppink','r'])
plt.title('算法时间成本分析图',size = 25)
plt.xlabel('解释算法',size = 20)
plt.ylabel('解释10个样本总时间成本（/s）',size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.rcParams['figure.figsize'] = (20,9)
plt.savefig("cost-pic.png")
```
