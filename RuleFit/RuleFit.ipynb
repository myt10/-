```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from rulefit import RuleFit
```

```python
df = pd.read_csv("../train/train_data_0.csv")
df
```

# exp1 线性生成Label,使用使用EBM Boosting解释

### 线性生成Label

$ y = 5 \times x_0 + 1 \times x_1 + 2 \times x_3$ 

```python
def normalize(x):
    x_max = x.max()
    x_min = x.min()
    if x.max() == x.min():
        return x
    else:
        result = (x-x_min)/(x_max-x_min)
        return result
df["label"] = df.feature0 *1 + df.feature1 * 1 + df.feature3*1
df["label"] = (normalize(df["label"])>0.5).astype(int)
df
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size = 0.3,random_state = 42)
```

```python
rf = RuleFit()
X_train = X_train.values()
rf.fit(X_train, y_train, feature_names=df.iloc[:,:-1].columns)
```

```python
rules = rf.get_rules()
rules
```

```python
rules = rules[rules.coef != 0].sort_values("importance", ascending=False)
rules
```
